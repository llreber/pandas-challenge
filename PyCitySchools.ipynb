```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
#school_data_complete.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = school_data_complete["school_name"].value_counts().count()
student_count = school_data_complete["student_name"].count()
school_budget = school_data_complete.groupby("school_name")["budget"].max()
total_budget = school_budget.sum()
student_budget = total_budget/student_count
ave_math= school_data_complete["math_score"].mean()
ave_read= school_data_complete["reading_score"].mean()
pass_math_pct = 100*school_data_complete[school_data_complete["math_score"]>=70].count()["math_score"]/student_count
pass_read_pct = 100*school_data_complete[school_data_complete["reading_score"]>=70].count()["reading_score"]/student_count
overall_pass_pct = (pass_math_pct+pass_read_pct)/2
school_totals = [school_count, student_count, total_budget, student_budget, ave_math, ave_read, pass_math_pct, pass_read_pct, overall_pass_pct]
school_totals
```

```python

school_headings = ["School Count","Student Count","Total Budget","Budget per Student","Average math Score","Average Reading Score","Math Passing %","Reading Passing %","Overall Pass Rate"]
school_summary = pd.DataFrame([school_totals], columns = school_headings)
school_summary
```
